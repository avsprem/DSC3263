Clean the Data Set by removing unnecessary columns, null rows and table headers

```python
df = pd.read_csv("output.csv",skiprows=2,header=None,encoding="IBM437")
df
```

```python


columns_to_drop = [0] + [3, 4] + list(range(13, len(df.columns)))

if len(df.columns) == 18:
    df.drop(columns=df.columns[columns_to_drop], inplace=True)
else:
    df.drop(df.columns[[0] + list(range(11, len(df.columns)))], axis=1, inplace=True)

df = df.dropna(thresh=5)
df.reset_index(drop=True, inplace=True)
```

Reading the rainfall Data into Pandas Series 

```python

reader = PdfReader(path)
number_of_pages = len(reader.pages)
page = reader.pages[0]
text = page.extract_text()

# text processing logic
extracted_values = []
for i in range(len(text.split("\n")[:-3])):
    if len(text.split("\n")[i]) > 40:
        extracted_values.append(text.split("\n")[i].split(" ")[-1])

# Create a Pandas Series
rf_data = pd.Series(extracted_values)

```

```python
rf_data
```

Convering Rainfall data from String into numeric type, cleaning and add to the datafame as a new column

```python
rf_data= rf_data[0:40]
rf_data = pd.to_numeric(rf_data, errors='coerce').fillna(0)
df = pd.concat([df, rf_data], axis=1)
```

Replace the "-" with 0 in numerical columns

```python
water_level_columns = ['Alert Level', 'Minor Flood Level', 'Major Flood Level', 'Water Level at 8:00 am', 'Water Level at 9:00 am']
df[water_level_columns] = df[water_level_columns].replace('-',pd.NA)
df[water_level_columns] = df[water_level_columns].fillna(0)
df
```

Convert the unit of Station "N' Street" from feet to meters

```python
row_condition = (df['Station'] == 'N\' Street')
df.loc[row_condition, 'Unit'] = 'm'

df[water_level_columns] = df[water_level_columns].astype(float)
df.loc[row_condition, water_level_columns] = (df.loc[row_condition, water_level_columns] * 0.3048).round(2)
```
