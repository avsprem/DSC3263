### Importing necessary Libraries
tabula - to read table in to pandas DF <br>
PyPDF2.pdfreader - to read rainfall data  

```python
import tabula
from PyPDF2 import PdfReader
import pandas as pd
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
import numpy as np
```

Link to the PDF

```python

path = "C:/Users/User/Desktop/Prac/1_DATA.pdf"
```

Reading table data into padas DF

```python
try:
    tabula.convert_into(path, "output.csv", output_format="csv", pages='1')
    print("converted")

except Exception as e:
    print("Error {}".format(e))

```

Clean the Data Set by removing unnecessary columns, null rows and table headers

```python
df = pd.read_csv("output.csv",skiprows=2,header=None)
df.drop(df.columns[[0] + list(range(11, len(df.columns)))], axis=1, inplace=True)
df = df.dropna(how='all')
df.reset_index(drop=True, inplace=True)
```

Cleaned Dataframe

```python
df
```

Reading the rainfall Data into Pandas Series 

```python

reader = PdfReader(path)
number_of_pages = len(reader.pages)
page = reader.pages[0]
text = page.extract_text()

# text processing logic
extracted_values = []
for i in range(len(text.split("\n")[:-3])):
    if len(text.split("\n")[i]) > 40:
        extracted_values.append(text.split("\n")[i].split(" ")[-1])

# Create a Pandas Series
rf_data = pd.Series(extracted_values)

```

```python
rf_data
```

Convering Rainfall data from String into numeric type, cleaning and add to the datafame as a new column

```python
rf_data= rf_data.drop(rf_data.index[-2:])
rf_data = pd.to_numeric(rf_data, errors='coerce').fillna(0)
df = pd.concat([df, rf_data], axis=1)
```

Rename the Dataframe Columns

```python
df.columns = ["Tributory/River", "Station", "Unit", "Alert Level", "Minor Flood Level", "Major Flood Level", "Water Level at 8:00 am", "Water Level at 9:00 am", "Remarks", "Water Level Rising or Falling", "24 Hr RF in mm at 8.30 am"]
```

```python
df["Tributory/River"]=df["Tributory/River"].fillna("Mahaweli Ganga")
df["Water Level Rising or Falling"]=df["Water Level Rising or Falling"].fillna("Stable")
```
